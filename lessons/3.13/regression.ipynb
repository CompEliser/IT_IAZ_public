```python
# Імпортування необхідних бібліотек
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Генерація датасету
X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=42)

# Конвертація в pandas DataFrame для зручності
data = pd.DataFrame(data=np.c_[X, y], columns=['X', 'y'])

# Візуалізація сгенерованих даних
plt.scatter(data['X'], data['y'])
plt.title("Сгенеровані дані")
plt.xlabel("Незалежна змінна X")
plt.ylabel("Залежна змінна y")
plt.show()

# Візуалізація матриці кореляції
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Матриця кореляції")
plt.show()

# Розділення даних на навчальний та тестовий набори
X_train, X_test, y_train, y_test = train_test_split(data['X'].values.reshape(-1,1), data['y'], test_size=0.2, random_state=42)

# Створення моделі лінійної регресії
model = LinearRegression()

# Навчання моделі на навчальному наборі
model.fit(X_train, y_train)

# Передбачення значень на тестовому наборі
y_pred = model.predict(X_test)

# Оцінка моделі
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Середньоквадратична помилка (MSE): {mse}")
print(f"Коефіцієнт детермінації (R^2): {r2}")

# Візуалізація результатів
plt.scatter(X_train, y_train, color='blue', label='Навчальні дані')
plt.scatter(X_test, y_test, color='green', label='Тестові дані')
plt.plot(X_test, y_pred, color='red', label='Лінія регресії')
plt.title("Лінійна регресія")
plt.xlabel("Незалежна змінна X")
plt.ylabel("Залежна змінна y")
plt.legend()
plt.show()
```

```python
# Імпортування необхідних бібліотек
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Генерація датасету з трьома предикторами
X, y = make_regression(n_samples=100, n_features=3, noise=10, random_state=42)

# Конвертація в pandas DataFrame для зручності
data = pd.DataFrame(data=X, columns=['X1', 'X2', 'X3'])
data['y'] = y

# Візуалізація матриці кореляції
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Матриця кореляції")
plt.show()

# Розділення даних на навчальний та тестовий набори
X_train, X_test, y_train, y_test = train_test_split(data[['X1', 'X2', 'X3']], data['y'], test_size=0.2, random_state=42)

# Створення моделі лінійної регресії
model = LinearRegression()

# Навчання моделі на навчальному наборі
model.fit(X_train, y_train)

# Передбачення значень на тестовому наборі
y_pred = model.predict(X_test)

# Оцінка моделі
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Середньоквадратична помилка (MSE): {mse}")
print(f"Коефіцієнт детермінації (R^2): {r2}")
```
