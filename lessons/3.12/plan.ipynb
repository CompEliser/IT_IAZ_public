1. **Вступ до штучного інтелекту та аналізу даних**

   - Загальний огляд застосування АІ в аналізі даних.
   - Основні поняття та термінологія.

2. **Типи даних та їх обробка**
   - Структуровані та неструктуровані дані.
   - Попередня обробка даних (очищення, нормалізація, перетворення).
   - Інструменти для обробки даних (Pandas, NumPy).

3. **Візуалізація даних**
   - Основи візуалізації (Matplotlib, Seaborn).
   - Інтерпретація візуальних даних.

1. **Основи машинного навчання**

2. **Вибір та оцінка моделей**
   - Критерії вибору алгоритмів.

3. **Практичні заняття з машинного навчання**
