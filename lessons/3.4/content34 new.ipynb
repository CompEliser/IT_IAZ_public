#### Створення бази даних та таблиці

```python
import psycopg2

# Підключення до PostgreSQL (замініть параметри на свої)
conn = psycopg2.connect(
    host="localhost",
    database="mydatabase",
    user="myuser",
    password="mypassword"
)

# Створення курсора для виконання SQL-запитів
cur = conn.cursor()

# Створення таблиці
create_table_query = '''
    CREATE TABLE IF NOT EXISTS users (
        id SERIAL PRIMARY KEY,
        name VARCHAR(255),
        age INT
    )
'''
cur.execute(create_table_query)
conn.commit()

# Закриття підключення
cur.close()
conn.close()
```

Розділ 2: Математична основа баз даних

#### Математична теорія множин у контексті баз даних

Математична теорія множин є важливим фундаментом для реляційних баз даних. Основні поняття теорії множин, які застосовуються у контексті баз даних, включають:

1. **Множини (Sets)**: Множина - це набір об'єктів. У базах даних множини відповідають таблицям, де кожен рядок таблиці представляє собою кортеж.

2. **Підмножина (Subset)**: Підмножина - це множина, яка складається з елементів більшої множини. У базах даних це використовується для операцій фільтрації даних.

3. **Об'єднання (Union)**: Об'єднання двох множин - це створення нової множини, яка містить всі унікальні елементи обох множин. У контексті баз даних це використовується для об'єднання результатів двох запитів.

4. **Перетин (Intersection)**: Перетин двох множин - це множина, яка містить всі спільні елементи обох множин. У базах даних це використовується для вибору даних, які належать до обох умов.

5. **Різниця (Difference)**: Різниця двох множин - це множина, яка містить елементи першої множини, які не належать до другої. У базах даних це використовується для виключення певних даних.

#### Відношення (relations) як основний математичний об'єкт у реляційних базах даних

В реляційних базах даних відношення є центральним поняттям. Відношення - це математична структура, яка відповідає таблиці в базі даних. Воно містить кортежі (рядки) і атрибути (стовпці).

- **Кортеж (Tuple)**: Кортеж - це запис в таблиці (рядок), який містить значення атрибутів для певного об'єкта чи сутності.

- **Атрибут (Attribute)**: Атрибут - це стовпець у відношенні, який визначає характеристики об'єктів у таблиці.

#### Операції над відношеннями: об'єднання, перетин, різниця

У реляційних базах даних можна виконувати різні операції над відношеннями:

1. **Об'єднання (Union)**: Операція об'єднання дозволяє об'єднати два відношення в одне, включаючи всі унікальні кортежі з обох відношень.

2. **Перетин (Intersection)**: Операція перетину видає кортежі, які належать до обох відношень, відкидаючи ті, які не є спільними для обох.

3. **Різниця (Difference)**: Операція різниці видає кортежі, які належать до першого відношення, але не належать до другого.

Ці операції допомагають вибирати, комбінувати та обробляти дані в базі даних з урахуванням відносин між ними.

Все це становить математичну основу реляційних баз даних і дозволяє вам ефективно управляти та опрацьовувати дані в системах управління базами даних.

Щоб продемонструвати операції над відношеннями в реляційних базах даних, давайте розглянемо приклад з використанням мови SQL і бази даних PostgreSQL.

1. Спершу, створімо дві таблиці з даними. Припустимо, ми маємо таблиці "Students" та "Courses" для зберігання інформації про студентів і курси, а також таблицю "Enrollments" для зберігання інформації про записи студентів на курси:

```sql
-- Створення таблиці "Students"
CREATE TABLE Students (
    student_id serial PRIMARY KEY,
    student_name VARCHAR(255) NOT NULL
);

-- Створення таблиці "Courses"
CREATE TABLE Courses (
    course_id serial PRIMARY KEY,
    course_name VARCHAR(255) NOT NULL
);

-- Створення таблиці "Enrollments" для зв'язку студентів і курсів
CREATE TABLE Enrollments (
    enrollment_id serial PRIMARY KEY,
    student_id INT REFERENCES Students(student_id),
    course_id INT REFERENCES Courses(course_id)
);
```

2. Заповнимо таблиці даними:

```sql
-- Додавання даних до таблиці "Students"
INSERT INTO Students (student_name) VALUES
    ('John'),
    ('Alice'),
    ('Bob');

-- Додавання даних до таблиці "Courses"
INSERT INTO Courses (course_name) VALUES
    ('Math'),
    ('History'),
    ('Computer Science');

-- Записи студентів на курси (Enrollments)
INSERT INTO Enrollments (student_id, course_id) VALUES
    (1, 1), -- John на курсі Math
    (1, 3), -- John на курсі Computer Science
    (2, 2), -- Alice на курсі History
    (3, 1); -- Bob на курсі Math
```

3. Тепер ми можемо виконати деякі операції над цими таблицями:

- Об'єднання (Union) студентів, які вивчають математику або історію:

```sql
SELECT student_name
FROM Students
WHERE student_id IN (
    SELECT student_id FROM Enrollments WHERE course_id = 1
    UNION
    SELECT student_id FROM Enrollments WHERE course_id = 2
);
```

- Перетин (Intersection) студентів, які вивчають математику і комп'ютерні науки:

```sql
SELECT student_name
FROM Students
WHERE student_id IN (
    SELECT student_id FROM Enrollments WHERE course_id = 1
    INTERSECT
    SELECT student_id FROM Enrollments WHERE course_id = 3
);
```

- Різниця (Difference) студентів, які вивчають історію, але не вивчають комп'ютерні науки:

```sql
SELECT student_name
FROM Students
WHERE student_id IN (
    SELECT student_id FROM Enrollments WHERE course_id = 2
    EXCEPT
    SELECT student_id FROM Enrollments WHERE course_id = 3
);
```

Це лише декілька прикладів операцій над відношеннями, які можна виконувати у реляційних базах даних. Операції об'єднання, перетину та різниці допомагають вибирати та фільтрувати дані з таблиць, що є важливими у роботі з базами даних.

#### Мережеві бази даних: призначення та особливості

**Мережева база даних** - це тип бази даних, яка може зберігати дані у вигляді зв'язків між записами, а не тільки в ієрархічній структурі. Основні особливості мережевих баз даних включають:

1. **Зв'язки**: В даній моделі зв'язки між записами можуть бути багаторазовими, а записи можуть мати кілька батьківських записів.

2. **Графова структура**: Дані в мережевій базі можуть бути представлені у вигляді графової структури з вузлами і ребрами.

3. **Призначення**: Мережеві бази даних використовуються там, де потрібно моделювати складні зв'язки між даними, такі як генеалогічні дерева, дерева шляхів файлових систем і ін.

#### Об'єктно-орієнтовані бази даних: впровадження об'єктно-орієнтованих концепцій у базах даних

**Об'єктно-орієнтована база даних (OODB)** - це тип бази даних, в якому дані організовані у вигляді об'єктів з методами та властивостями. Основні особливості об'єктно-орієнтованих баз даних включають:

1. **Об'єктна модель**: Дані представлені у вигляді об'єктів з властивостями та методами.

3. **Успадкування**: Успадкування є важливою особливістю об'єктно-орієнтованих баз даних. Об'єкти можуть успадковувати властивості та методи від інших об'єктів, що спрощує організацію та використання коду. Це використовується для створення ієрархій об'єктів, де підкласи успадковують властивості та методи від батьківських класів.

4. **Асоціації**: В об'єктно-орієнтованих базах даних можуть існувати асоціації між об'єктами. Це дозволяє визначати взаємні зв'язки між об'єктами, що відображають реальні відносини між даними. Наприклад, об'єкт "студент" може бути асоційований з об'єктом "курс", вказуючи, які студенти зареєстровані на певний курс.

5. **Поліморфізм**: Об'єктно-орієнтовані бази даних підтримують поліморфізм, що дозволяє об'єктам реагувати на запити або операції специфічним для них способом. Це робить код більш гнучким та відділеною від типу даних.

6. **Зберігання об'єктів**: Однією з важливих особливостей OODB є здатність зберігати об'єкти у базі даних без необхідності перетворення їх у реляційну структуру. Це дозволяє зберігати та використовувати об'єкти у їхньому природному стані.

7. **Запити та мови запитів**: В OODB використовуються спеціалізовані мови запитів, які дозволяють виконувати операції з об'єктами та їхніми властивостями. Зазвичай це мови запитів, які схожі на SQL, але адаптовані для роботи з об'єктами.

8. **Інкапсуляція та приховування даних**: Об'єктно-орієнтовані бази даних сприяють інкапсуляції даних та приховуванню їхньої реалізації від користувачів. Це підвищує безпеку та дозволяє управляти доступом до даних.

#### Ієрархічні бази даних

У цьому прикладі ми використовуємо деревоподібну структуру даних для моделювання організаційної структури компанії:

```python
class Employee:
    def __init__(self, name, position):
        self.name = name
        self.position = position
        self.subordinates = []

    def add_subordinate(self, employee):
        self.subordinates.append(employee)

# Створення ієрархії співробітників
ceo = Employee("John", "CEO")
cto = Employee("Alice", "CTO")
hr_manager = Employee("Bob", "HR Manager")

ceo.add_subordinate(cto)
ceo.add_subordinate(hr_manager)

developer = Employee("Eva", "Developer")
designer = Employee("Mike", "Designer")

cto.add_subordinate(developer)
cto.add_subordinate(designer)

# Виведення ієрархії на екран
def print_hierarchy(employee, level=0):
    print("  " * level + f"{employee.name} ({employee.position})")
    for subordinate in employee.subordinates:
        print_hierarchy(subordinate, level + 1)

print_hierarchy(ceo)
```

#### Мережеві бази даних

У мережевих базах даних, зазвичай, використовуються спеціальні системи керування базами даних, такі як Neo4j. Ось приклад створення графової структури за допомогою Neo4j та мови Python:

```python
from neo4j import GraphDatabase

# Підключення до бази даних Neo4j
uri = "bolt://localhost:7687"
username = "neo4j"
password = "password"

class Employee:
    def __init__(self, name, position):
        self.name = name
        self.position = position

def add_employee(tx, employee, manager=None):
    if manager:
        query = (
            "MATCH (manager:Employee {name: $manager_name})"
            "CREATE (employee:Employee {name: $name, position: $position})"
            "MERGE (employee)-[:REPORTS_TO]->(manager)"
        )
        tx.run(query, manager_name=manager.name, name=employee.name, position=employee.position)
    else:
        query = (
            "CREATE (employee:Employee {name: $name, position: $position})"
        )
        tx.run(query, name=employee.name, position=employee.position)

# Створення ієрархії співробітників у Neo4j
with GraphDatabase.driver(uri, auth=(username, password)) as driver:
    with driver.session() as session:
        ceo = Employee("John", "CEO")
        cto = Employee("Alice", "CTO")
        hr_manager = Employee("Bob", "HR Manager")

        session.write_transaction(add_employee, ceo)
        session.write_transaction(add_employee, cto, manager=ceo)
        session.write_transaction(add_employee, hr_manager, manager=ceo)

        developer = Employee("Eva", "Developer")
        designer = Employee("Mike", "Designer")

        session.write_transaction(add_employee, developer, manager=cto)
        session.write_transaction(add_employee, designer, manager=cto)
```

#### Об'єктно-орієнтовані бази даних (OODB)

Python має багато бібліотек для роботи з об'єктно-орієнтованими базами даних. Один з прикладів - бібліотека ZODB. Вот приклад створення та роботи з об'єктною базою даних ZODB:

```python
import ZODB
import ZODB.FileStorage
import transaction

class Employee:
    def __init__(self, name, position):
        self.name = name
        self.position = position
        self.subordinates = []

# Створення об'єктної бази даних
storage = ZODB.FileStorage.FileStorage('mydb.fs')
db = ZODB.DB(storage)
connection = db.open()
root = connection.root()

# Додавання об'єктів до бази даних
ceo = Employee("John", "CEO")
cto = Employee("Alice", "CTO")
hr_manager = Employee("Bob", "HR Manager")

ceo.subordinates.extend([cto, hr_manager])

root['ceo'] = ceo

# Збереження змін до бази даних
transaction.commit()

# Отримання даних з бази даних
loaded_ceo = root['ceo']
print(loaded_ceo.name, loaded_ceo.position)

# Закриття підключення до бази даних
connection.close()
storage.close()
```

   - Атрибут - це конкретна характеристика або властивість сутності, яка містить конкретну інформацію.
   - Атрибути допомагають деталізувати сутності та роблять дані в базі даних більш структурованими та корисними.

   - Реляційна модель дозволяє легко встановлювати відносини між сутностями та виконувати операції над даними за допомогою SQL-запитів.

   - Назва таблиці повинна бути унікальною у межах бази даних.

2. **Записи в таблицях**

   - Запис (або кортеж) - це конкретний рядок у таблиці, який містить дані для певної сутності.
   - Кожен запис містить значення атрибутів, які визначають характеристики сутності.

#### Ключі (Primary Key, Foreign Key)

1. **Первинний ключ (Primary Key)**

   - Первинний ключ - це унікальний ідентифікатор для записів в таблиці.
   - Він гарантує, що кожен запис має унікальний ідентифікатор, який можна використовувати для пошуку та звернення до запису.
   - Зазвичай, первинний ключ створюється за допомогою одного або кількох атрибутів.

2. **Зовнішній ключ (Foreign Key)**

   - Зовнішній ключ - це атрибут, який посилається на первинний ключ іншої таблиці.
   - Він встановлює зв'язок між таблицями та дозволяє виконувати операції з'єднання (JOIN) для отримання даних з різних таблиць.

### Розділ 4: Нормалізація даних

#### Поняття нормалізації та її мета

1. **Нормалізація даних**

   - Нормалізація даних - це процес організації даних в базі даних з метою забезпечення їхньої ефективної та коректної обробки.
   - Головною метою нормалізації є усунення аномалій даних та забезпечення цілісності даних.

#### Перша, друга, третя та інші нормальні форми

2. **Перша нормальна форма (1NF)**

   - Для досягнення 1NF всі значення у таблиці повинні бути атомарними (недільними) та не розділятися на підзначення.
   - Кожен стовпець повинен містити атомарні значення.
   
3. **Друга нормальна форма (2NF)**

   - Для досягнення 2NF, таблиці повинні вже бути в 1NF, і кожен стовпець, що не є частиною первинного ключа, повинен залежати від всього первинного ключа, а не від часткового.
   - Це означає, що в таблиці не повинно бути часткової залежності від первинного ключа.

4. **Третя нормальна форма (3NF)**

   - Для досягнення 3NF, таблиці повинні вже бути в 2NF, і всі неключові стовпці повинні бути непрямо залежні від первинного ключа, тобто вони не повинні залежати від інших неключових стовпців.
   - Це допомагає уникнути транзитивної залежності між стовпцями.

5. **Інші нормальні форми (BCNF, 4NF, 5NF, тощо)**

   - Існують інші нормальні форми, які розширюють концепції 1NF, 2NF та 3NF, і допомагають уникнути складних залежностей у таблицях.
   - BCNF (Boyce-Codd Normal Form), 4NF (Fourth Normal Form), 5NF (Fifth Normal Form) та інші форми використовуються для більш складних сценаріїв нормалізації.

#### Проблеми з ненормалізованими даними та їх вирішення

6. **Проблеми з ненормалізованими даними та їх вирішення**

   - Ненормалізовані дані можуть призводити до аномалій, таких як дублювання даних, втрата цілісності та складність обробки.
   - Нормалізація даних допомагає вирішити ці проблеми та полегшує підтримку та розвиток бази даних.

Цей матеріал надає загальний огляд нормалізації даних, включаючи поняття, мету та основні нормальні форми. Він також підкреслює проблеми, які можуть виникнути з ненормалізованими даними та переваги вирішення цих проблем шляхом нормалізації.

### Розділ 5: Транзакції та керування конкурентним доступом

#### Поняття транзакції

1. **Транзакція**

   - Транзакція - це логічна одиниця роботи з базою даних, яка включає один чи декілька SQL-запитів.
   - Транзакції використовуються для групування операцій та забезпечення їх атомарності та цілісності.

#### Властивості транзакцій: ACID

2. **ACID (Atomicity, Consistency, Isolation, Durability)**

   - ACID - це набір властивостей, які гарантують надійність операцій в транзакціях.
   - **Atomicity (Атомарність)**: Транзакція виконується атомарно, тобто вона або повністю виконується, або не виконується взагалі.
   - **Consistency (Цілісність)**: Транзакція переводить базу даних з одного стану в інший, який відповідає правильності та цілісності даних.
   - **Isolation (Ізоляція)**: Транзакції виконуються ізольовано одна від одної, інші транзакції не бачать проміжних результатів.
   - **Durability (Тривалість)**: Після успішного виконання транзакції її зміни залишаються незмінними навіть у випадку відмови системи.

#### Конкурентний доступ до бази даних та блокування

3. **Конкурентний доступ до бази даних**

   - Конкурентний доступ - це одночасний доступ кількох користувачів до бази даних.
   - Керування конкурентним доступом важливо для забезпечення цілісності та надійності даних.

4. **Блокування**

   - Блокування - це механізм, який використовується для контролю доступу до ресурсів бази даних у многокористувацькому середовищі.
   - Блокування дозволяє одній транзакції блокувати ресурс для інших транзакцій до завершення своєї роботи.

Цей матеріал надає загальний огляд транзакцій, їхніх властивостей (ACID) та конкурентного доступу до бази даних з використанням блокування. Він допоможе вам зрозуміти важливість правильного керування транзакціями та контролю над доступом до даних у базі даних.

Ось приклад використання транзакцій та блокування в PostgreSQL (але аналогічний підхід можна застосувати до інших систем керування базами даних):

1. Спершу, створіть таблицю "Рахунки":

```sql
CREATE TABLE Accounts (
    account_id serial PRIMARY KEY,
    account_name VARCHAR(255) NOT NULL,
    balance DECIMAL(10, 2) NOT NULL
);

INSERT INTO Accounts (account_name, balance) VALUES
    ('Рахунок 1', 1000.00),
    ('Рахунок 2', 500.00);
```

2. Тепер, давайте розглянемо приклад переказу коштів між рахунками з використанням транзакцій та блокування:

```sql
BEGIN; -- Початок транзакції

-- Отримуємо баланс рахунку 1
SELECT balance FROM Accounts WHERE account_name = 'Рахунок 1' FOR UPDATE;

-- Отримуємо баланс рахунку 2
SELECT balance FROM Accounts WHERE account_name = 'Рахунок 2' FOR UPDATE;

-- Переказ коштів із Рахунку 1 на Рахунок 2
UPDATE Accounts SET balance = balance - 200 WHERE account_name = 'Рахунок 1';
UPDATE Accounts SET balance = balance + 200 WHERE account_name = 'Рахунок 2';

COMMIT; -- Завершення транзакції
```

У цьому прикладі ми використовуємо команду `BEGIN` для початку транзакції і команду `COMMIT` для її завершення. За допомогою `FOR UPDATE` ми заблоковуємо записи, щоб уникнути конфліктів при одночасному доступі до рахунків.

Цей приклад демонструє використання транзакцій та блокування для забезпечення консистентності даних під час переказу коштів між рахунками.
