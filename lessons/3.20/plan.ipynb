### Основи використання ІТ у веб-дизайні

**1. ІТ розмітки веб-сторінки (45 хвилин)**

**Вступ до ІТ розмітки веб-сторінок**
- Роз'яснення, що таке ІТ розмітка в контексті веб-дизайну.
- Огляд основних технологій розмітки, зокрема HTML.
- Розгляд структури HTML-документа, включаючи елементи, атрибути та ієрархію.

**Практична робота з HTML**
- Демонстрація створення базової структури веб-сторінки з використанням HTML.
- Вправи на створення різних HTML-елементів (заголовки, параграфи, списки, зображення, посилання тощо).
- Обговорення важливості семантичної розмітки та доступності веб-сторінок.

**2. ІТ поведінки веб-сторінки (45 хвилин)**

**Вступ до ІТ поведінки веб-сторінок**
- Роз'яснення концепції ІТ поведінки в контексті веб-дизайну.
- Введення в JavaScript як основний інструмент для контролю поведінки веб-сторінки.
- Основні концепції JavaScript: змінні, функції, об'єкти, події.

**Практична робота з JavaScript**
- Демонстрація інтеграції JavaScript у веб-сторінку для додавання динамічної поведінки.
- Вправи на створення простих скриптів JavaScript для обробки подій (наприклад, натискання кнопки).
- Обговорення впливу JavaScript на користувацький досвід та інтерактивність веб-сторінки.

**Заключення**
- Підведення підсумків та обговорення важливості інтеграції ІТ у веб-дизайн.
- Відповіді на запитання.
- Рекомендації для подальшого самостійного вивчення.

Цей план забезпечує цілісне охоплення основних аспектів використання ІТ у веб-дизайні, з особливим акцентом на розмітку та поведінку веб-сторінок. Він включає як теоретичні основи, так і практичні вправи, що дозволяє учасникам заняття активно застосовувати отримані знання.
