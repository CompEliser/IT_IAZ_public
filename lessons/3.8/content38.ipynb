Структура мови SQL (Structured Query Language) може бути розділена на декілька основних компонентів, кожен з яких відіграє унікальну роль у управлінні та маніпуляції даними в реляційних базах даних:

### 1. Data Definition Language (DDL)
- **Завдання**: Контролює структури та схеми баз даних.
- **Ключові команди**:
  - `CREATE`: Створює нові таблиці, бази даних, індекси та інші об'єкти.
  - `ALTER`: Модифікує існуючі об'єкти бази даних.
  - `DROP`: Видаляє об'єкти з бази даних.

### 2. Data Manipulation Language (DML)
- **Завдання**: Управління даними всередині таблиць.
- **Ключові команди**:
  - `SELECT`: Використовується для вибору даних з таблиці.
  - `INSERT`: Додає нові рядки до таблиці.
  - `UPDATE`: Оновлює існуючі дані в таблиці.
  - `DELETE`: Видаляє рядки з таблиці.

### 3. Data Control Language (DCL)
- **Завдання**: Управління правами доступу до даних.
- **Ключові команди**:
  - `GRANT`: Надає користувачам права доступу до об'єктів бази даних.
  - `REVOKE`: Відкликає раніше надані права доступу.

### 4. Transaction Control Language (TCL)
- **Завдання**: Управління транзакціями в базі даних.
- **Ключові команди**:
  - `COMMIT`: Застосовує зміни, зроблені під час транзакції.
  - `ROLLBACK`: Відкочує зміни до останнього COMMIT.
  - `SAVEPOINT`: Встановлює точки відновлення у межах транзакції.

### 5. Data Query Language (DQL)
- **Завдання**: Виключно використовується для запитів до даних (часто розглядається як частина DML).
- **Основна команда**:
  - `SELECT`: Використовується для запиту даних з бази.

Кожен з цих компонентів виконує специфічні завдання, які разом дозволяють ефективно управляти даними, структурою бази даних, транзакціями, та доступом до даних у реляційних базах даних.

### Контент навчального матеріалу: Огляд Data Manipulation Language (DML)

#### Вступ до DML
- **Визначення**: Data Manipulation Language (DML) є частиною SQL, яка використовується для вибору, вставки, оновлення та видалення даних у базах даних.
- **Роль у роботі з базами даних**: DML надає користувачам та розробникам інструменти для управління даними, від простих запитів до складних аналітичних операцій.

#### Основні команди DML
1. **SELECT**
   - Використовується для вибору даних з однієї або декількох таблиць.
   - Підтримує різноманітні опції фільтрації, сортування та агрегації.
2. **INSERT**
   - Додає нові записи до таблиці.
   - Може вставляти один рядок за раз або декілька рядків одночасно.
3. **UPDATE**
   - Змінює існуючі записи в таблиці.
   - Потребує критеріїв вибору для визначення, які саме рядки оновлювати.
4. **DELETE**
   - Видаляє записи з таблиці.
   - Також потребує критеріїв для визначення видаляємих рядків.

#### Розширені можливості DML
- **JOINs**: Дозволяє об'єднувати дані з різних таблиць на основі спільних стовпців.
- **Subqueries**: Запит в межах іншого запиту, що дозволяє виконувати більш складні операції вибору.
- **Set Operations**: Операції, як UNION, INTERSECT, EXCEPT, які дозволяють об'єднувати результати декількох запитів.

#### Заключення
- **Значення DML**: Посилює гнучкість та ефективність в роботі з базами даних, дозволяючи розробникам та аналітикам виконувати широкий спектр операцій з даними.
- **Безперервний розвиток**: Сучасні бази даних продовжують розширювати можливості DML, включаючи підтримку складних аналітичних та транзакційних операцій.

### Контент для "Основні концепції DML SQL"

#### Основні концепції DML SQL
1. **Що таке DML**: Data Manipulation Language (DML) - це частина SQL, яка використовується для вставки, зміни, оновлення та видалення даних у базі даних.
2. **Взаємодія з даними**: DML дозволяє користувачам здійснювати операції з даними, які вже знаходяться в таблицях.
3. **Транзакційність**: Більшість DML-операцій можуть бути транзакційними, що означає, що зміни можуть бути відкочені назад (ROLLBACK) або збережені (COMMIT).

#### Перегляд ключових команд DML
1. **SELECT**
   - Використовується для вибору та виведення даних з однієї або декількох таблиць.
   - Підтримує фільтрацію, сортування, об'єднання таблиць та агрегування даних.
2. **INSERT**
   - Дозволяє додавати нові рядки до таблиці.
   - Можна вставляти як один рядок, так і декілька рядків одночасно.
3. **UPDATE**
   - Змінює існуючі рядки в таблиці.
   - Потребує критеріїв вибору (WHERE clause) для визначення, які саме рядки оновлювати.
4. **DELETE**
   - Видаляє рядки з таблиці.
   - Також вимагає критеріїв вибору для визначення, які рядки видалити.

#### Принципи побудови ефективних SQL-запитів
1. **Оптимізація вибору**: Використовуйте конкретні умови в WHERE clause, щоб обмежити кількість рядків, які обробляються.
2. **Індексація**: Використання індексів може значно прискорити запити, особливо при великій кількості даних.
3. **Уникнення непотрібних JOIN-ів**: Зайві об'єднання таблиць можуть сповільнити запит.
4. **Використання агрегатних функцій**: Для обробки даних на стороні сервера, а не на клієнті.
5. **Планування запитів**: Аналізуйте плани виконання запитів, щоб зрозуміти, як оптимізувати їх виконання.
6. **Чистота коду**: Чітке форматування і коментування SQL-коду спрощує його розуміння та підтримку.

### Контент для "Використання агрегатних функцій у SQL"

#### Вступ до агрегатних функцій
- **Визначення**: Агрегатні функції в SQL використовуються для виконання обчислень на наборі значень і повернення єдиного значення.
- **Застосування**: Широко використовуються для аналізу та зведення даних в базах даних.

#### Основні агрегатні функції
1. **SUM()**
   - Використовується для підсумовування числових значень в колонці.
   - Приклад: `SELECT SUM(salary) FROM employees;`

2. **AVG()**
   - Обчислює середнє значення числових даних.
   - Приклад: `SELECT AVG(salary) FROM employees;`

3. **COUNT()**
   - Визначає кількість рядків, які відповідають вказаним критеріям.
   - Приклад: `SELECT COUNT(*) FROM employees WHERE department = 'Sales';`

4. **MAX()**
   - Повертає максимальне значення в колонці.
   - Приклад: `SELECT MAX(salary) FROM employees;`

5. **MIN()**
   - Повертає мінімальне значення в колонці.
   - Приклад: `SELECT MIN(salary) FROM employees;`

#### Групування даних з GROUP BY
- **Застосування**: Агрегатні функції часто використовуються разом із `GROUP BY` для групування значень у різних рядках.
- Приклад: `SELECT department, AVG(salary) FROM employees GROUP BY department;`

#### Використання HAVING для фільтрації агрегованих даних
- **Призначення**: `HAVING` використовується для фільтрації результатів, отриманих з агрегатних функцій.
- Приклад: `SELECT department, SUM(salary) FROM employees GROUP BY department HAVING SUM(salary) > 100000;`

#### Важливість агрегатних функцій
- **Аналітичні можливості**: Агрегатні функції є ключовими для аналізу даних, дозволяючи швидко обробляти великі обсяги інформації.
- **Ефективність**: Вони дозволяють ефективно виконувати обчислення на сервері бази даних, зменшуючи навантаження на мережу та клієнтські застосунки.

### Контент для "Window Functions у SQL"

#### Поняття та Класифікація Window Functions
- **Визначення**: Window Functions в SQL - це функції, які надають унікальні можливості для виконання обчислень по набору рядків, що відносяться до поточного рядка.
- **Класифікація**:
  - **Агрегатні Window Functions**: Функції, які використовуються для агрегування даних (наприклад, `SUM()`, `AVG()`).
  - **Аналітичні Window Functions**: Функції, що виконують обчислення на основі відношення між рядками (наприклад, `ROW_NUMBER()`, `RANK()`).
  - **Операції з рядками**: Функції, які працюють з наступними чи попередніми рядками (наприклад, `LEAD()`, `LAG()`).

#### Основні Window Functions
1. **RANK()**
   - Призначення: Повертає ранг кожного рядка в групі, з дублюванням номерів рангу і пропусками.
   - Приклад: `RANK() OVER (ORDER BY score DESC)`

2. **ROW_NUMBER()**
   - Призначення: Присвоює унікальний номер кожному рядку в залежності від порядку сортування.
   - Приклад: `ROW_NUMBER() OVER (ORDER BY name)`

3. **LEAD()**
   - Призначення: Повертає значення наступного рядка у вікні.
   - Приклад: `LEAD(price) OVER (ORDER BY date)`

4. **LAG()**
   - Призначення: Повертає значення попереднього рядка у вікні.
   - Приклад: `LAG(price) OVER (ORDER BY date)`

5. **FIRST_VALUE()**
   - Призначення: Повертає перше значення в вікні.
   - Приклад: `FIRST_VALUE(salary) OVER (ORDER BY hire_date)`

6. **LAST_VALUE()**
   - Призначення: Повертає останнє значення в вікні.
   - Приклад: `LAST_VALUE(salary) OVER (ORDER BY hire_date)`

#### Розуміння OVER() клавзи
- **Основа для Window Functions**: `OVER()` клавза визначає, як буде сформовано "вікно" рядків для виконання обчислення.
- **Варіанти використання**:
  - **Без PARTITION BY**: Всі рядки вважаються частиною одного великого вікна.
  - **З PARTITION BY**: Розділяє дані на окремі "вікна" для кожної групи, вказаної в PARTITION BY.
  - **З ORDER BY**: Визначає порядок рядків у кожному вікні для обчислень.
- **Примітка**: `OVER()` клавза не змінює кількість рядків у результаті запиту; вона лише змінює те, як виконуються обчислення на цих рядках.

Для створення навчальної бази даних на тему укомплектованості підрозділів матеріальними засобами, ви можете використовувати такий SQL код для створення необхідних таблиць:

### SQL Скрипт для Створення Таблиць

```sql
-- Створення таблиці 'departments' (підрозділи)
CREATE TABLE departments (
    department_id INT PRIMARY KEY,
    name VARCHAR(255) NOT NULL,
    location VARCHAR(255)
);

-- Створення таблиці 'equipment_groups' (групи засобів)
CREATE TABLE equipment_groups (
    group_id INT PRIMARY KEY,
    group_name VARCHAR(255) NOT NULL
);

-- Створення таблиці 'equipments' (засоби)
CREATE TABLE equipments (
    equipment_id INT PRIMARY KEY,
    name VARCHAR(255) NOT NULL,
    group_id INT,
    FOREIGN KEY (group_id) REFERENCES equipment_groups(group_id)
);

-- Створення таблиці 'equipment_availability_standard' (наявність засобів за штатом)
CREATE TABLE equipment_availability_standard (
    standard_id INT PRIMARY KEY,
    department_id INT,
    equipment_id INT,
    quantity INT NOT NULL,
    FOREIGN KEY (department_id) REFERENCES departments(department_id),
    FOREIGN KEY (equipment_id) REFERENCES equipments(equipment_id)
);

-- Створення таблиці 'equipment_availability_list' (наявність засобів засписком)
CREATE TABLE equipment_availability_list (
    list_id INT PRIMARY KEY,
    department_id INT,
    equipment_id INT,
    quantity INT NOT NULL,
    condition VARCHAR(255),
    FOREIGN KEY (department_id) REFERENCES departments(department_id),
    FOREIGN KEY (equipment_id) REFERENCES equipments(equipment_id)
);
```

### Пояснення
1. **Таблиця `departments`**: Зберігає інформацію про підрозділи, включаючи їх ID, назву та місцезнаходження.
2. **Таблиця `equipment_groups`**: Визначає групи засобів, кожна з яких має унікальний ID та назву.
3. **Таблиця `equipments`**: Включає деталі конкретних засобів, такі як ID, назва та група, до якої вони належать.
4. **Таблиця `equipment_availability_standard`**: Передбачає штатне розташування засобів для кожного підрозділу.
5. **Таблиця `equipment_availability_list`**: Фіксує фактичну наявність засобів у підрозділах, включаючи їх стан.

Цей код створює основу для бази даних, яка може бути використана для управління та аналізу укомплектованості підрозділів матеріальними засобами.
