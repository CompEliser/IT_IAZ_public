Для заняття з майбутніми магістрами з дисципліни "Інформаційно-аналітичні технології", особливо з акцентом на використанні SQL для вибору інформації з бази даних, можна запропонувати таку структуру:

### Структура заняття (90 хвилин)

1. **Вступ (10 хвилин)**
   - Привітання та введення в тему.
   - Короткий огляд Data Manipulation Language (DML) і його ролі у роботі з базами даних.

2. **Теоретичні основи (20 хвилин)**
   - Основні концепції DML SQL.
   - Перегляд ключових команд (SELECT, INSERT, UPDATE, DELETE).
   - Принципи побудови ефективних SQL-запитів.
3. **Теоретичні основи Window Functions (20 хвилин)**
    - Поняття та класифікація Window Functions у SQL.
    - Основні функції: RANK, ROW_NUMBER, LEAD, LAG, FIRST_VALUE, LAST_VALUE.
    - Розуміння OVER() клавзи та її варіантів використання.

4. **Практична демонстрація (30 хвилин)**
   - Демонстрація використання PostgreSQL для вибору даних.
   - Приклади простих та складних SELECT запитів.
   - Показ використання JOIN, WHERE, GROUP BY та інших операторів.

5. **Практичні вправи (10 хвилин)**
   - Студенти самостійно пишуть SQL-запити на основі заданих умов.
   - Викладач консультує та надає зворотній зв'язок.
