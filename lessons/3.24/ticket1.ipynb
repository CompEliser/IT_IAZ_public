```python
import random

def assign_tickets_to_students(num_students, num_tickets, seed=None):
    if seed is not None:
        random.seed(seed)

    assigned_tickets = {}
    for student in range(1, num_students + 1):
        ticket_number = random.randint(1, num_tickets)
        assigned_tickets[f"Слухач {student}"] = ticket_number

    return assigned_tickets

# Приклад використання
num_students = 3  # Кількість слухачів
num_tickets = 22  # Кількість білетів
seed = 37857  # Ініціалізація генератора випадкових чисел

assigned_tickets = assign_tickets_to_students(num_students, num_tickets, seed)

unique_ticket_numbers = []

for student, ticket in assigned_tickets.items():
    print(f"{student} отримав білет №{ticket}")
    unique_ticket_numbers.append(ticket-1)

unique_ticket_numbers
```
