Структура 90-хвилинної лекції з теми "Основи вибору інформації з баз даних. Поняття про нереляційні бази даних" для майбутніх магістрів з дисципліни Інформаційно-аналітичні технології може виглядати наступним чином:

### 1. Вступ (10 хвилин)
   - Привітання та короткий огляд теми.
   - Значення баз даних у сучасному аналітичному світі.

### 2. Основи використання засобів вибору інформації з бази даних (30 хвилин)
   - Огляд SQL та його ролі у вибору даних (10 хвилин).
   - Практичне заняття: Основні команди SQL у Postgres, включаючи SELECT, JOIN, WHERE, GROUP BY (20 хвилин).

### 3. Поняття про нереляційні бази даних (20 хвилин)
   - Вступ до нереляційних баз даних: визначення та основні характеристики.
   - Різниця між реляційними та нереляційними базами даних.

### 4. Практична демонстрація (20 хвилин)
   - Демонстрація використання нереляційної бази даних на прикладі популярної системи (наприклад, MongoDB або Cassandra).
   - Порівняння запитів SQL та нереляційних запитів.

### 5. Відповіді на запитання та обговорення (10 хвилин)
   - Відповіді на запитання студентів.
   - Обговорення основних моментів лекції.

### Додатково:
   - Надання ресурсів для самостійного вивчення: книги, онлайн-курси, статті.
   - Коротке опитування для оцінки засвоєння матеріалу та зворотнього зв'язку.

Ця структура дозволить ефективно охопити ключові аспекти теми, поєднуючи теоретичні знання з практичними навичками.
